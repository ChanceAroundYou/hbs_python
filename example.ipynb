```python
from hbs.boundary import get_boundary, plot_boundary, smooth_resample
from hbs import get_hbs, reconstruct_from_hbs

img_path = 'img/example.jpg'
circle_point_num = 1000
density = 0.01
bound_point_num = 250
```

```python
bound = get_boundary(img_path, bound_point_num)
hbs, he, cw, disk = get_hbs(bound, circle_point_num, density)
cw.plot()
disk.plot_mu(hbs)
disk.plot_mapping(he)
```

```python
reconstructed_bound, _, _, reconstructed_he = reconstruct_from_hbs(hbs, disk)
disk.plot_mapping(reconstructed_he)
```

```python
plot_boundary(bound, scale=1)
plot_boundary(smooth_resample(reconstructed_bound, bound_point_num), scale=1)
# plot_boundary(reconstructed_bound, scale=1)
```
